```python
#Import
from ipywidgets import interact, Dropdown, Checkbox
import ipywidgets as widgets
from time import sleep
#Start
#Name
Name = input("Wie ist dein Name?:")
sleep (1)
print("Hallo",Name,"! Schön dich kennen zu lernen!")
#Frage
sleep (2)
print ("Welche Frage liegt dir auf dem Herzen?")
sleep (2)
Frage = input ("Deine Frage:")
Frage = Frage
print ("Das ist eine gute Frage..")
sleep (3)
print ("Lass mich mal überlegen.")
#Alter
sleep (3)
print ("Um dir eine genaue Antwort geben zu können, muss ich noch dein Alter wissen.")
sleep(1)
Alter = input ("Dein Alter:")
print ("Du bist also", Alter, "Jahre alt.")
sleep (3)
print ("Interessant..")
#Bedingung Alter
if int(Alter)<15:
    sleep(2)
    print("Du bist ja ganz schön jung. Vielleicht fragst du lieber deine Eltern nach einem Ratschlag. Denkst du nicht?")
    sleep (2)
else:
    sleep(2)
    print ("Ich denke du bist alt genug, deine eigenen Entscheidungen zu treffen. Denkst du nicht?")
#Radiobuttons Entscheidung
sleep(2)
radio = widgets.RadioButtons(options=['Wähle aus:','Ja, da hast du Recht allwissendes Orakel!', 'Nein, ich möchte lieber dich fragen, oh allmächtiges Orakel!'], layout={'width': 'max-content'}, value='Wähle aus:' )
widgets.Box(
    [widgets.Label(value='Was denkst du?'),
    radio
    ] )
text = ''
display(radio)
def eval(a):
    if  a == 'Nein, ich möchte lieber dich fragen, oh allmächtiges Orakel!':
        print ('Gut, ich weiß es sowieso am besten!')
    elif a == 'Wähle aus:':
        sleep (4)
        print ('Hallo?! Bekomme ich keine Antwort?')
    else:
        print('Vielen Dank für nichts! Dann hättest du mich ja gar nicht erst belästigen müssen!')
        sleep(2)
        print('Das mindeste was du jetzt tun kannst, ist mich noch ein wenig weiter zu bespaßen.')
out = widgets.interactive_output(eval, {'a': radio})
display(out)
```
